```python
import whisper

# Carregue o modelo (escolha entre 'tiny', 'base', 'small', 'medium', 'large')
model = whisper.load_model("medium")  # Recomendado para português

# Transcreva o áudio
result = model.transcribe("/content/drive/MyDrive/15_de_abr.m4a", language="pt")
print(result["text"])

# Salve a transcrição em um arquivo .txt
with open("transcricao.txt", "w", encoding="utf-8") as f:
    f.write(result["text"])
```
